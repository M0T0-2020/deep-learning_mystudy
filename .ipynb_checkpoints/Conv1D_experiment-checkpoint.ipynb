```python
import torch 
from torch import nn
from torch import optim
import matplotlib.pyplot as plt
import numpy as np
import pandas  as pd
import random

from tqdm import tqdm_notebook as tqdm
```

```python
def make_data(n):
    _data = []
    _true = []
    for _ in range(n):
        i = np.random.randint(-10, 10)
        x = np.linspace(i,i+6,200)
        x1 = np.sin(4*x)+np.random.rand(200)*0.3
        x2 = .3*(x**2)+np.random.rand(200)*0.3
        x3 = np.cos(5*x)+np.random.rand(200)*0.3
        true = x1+x2+x3
        x1 = x1.reshape(-1,1)
        x2 = x2.reshape(-1,1)
        x3 = x3.reshape(-1,1)
        x = np.hstack((x1,x2, x3))
        _data.append(x.tolist())
        _true.append(true)
    _data = torch.FloatTensor(_data)
    _true = torch.FloatTensor(_true)
    return _data, _true
```

```python
class Autoencorder_1d(nn.Module):

    def __init__(self, input_size, dim_num):
        super(Autoencorder_1d, self).__init__()
        
        size_1 = input_size
        size_2 = int(size_1/2.5) if int(size_1/2.5)<1 else 20
        size_3 = int(size_1/5) if int(size_1/5)<1 else 10
        
        
        self.encorder = nn.Sequential(
            nn.Conv1d(size_1, size_2, 2),
            nn.GELU(),
            nn.Conv1d(size_2, size_3, 2)
        )
        
        self.lin = nn.Linear(size_3, dim_num)
        
        self.avgpool1d = nn.AdaptiveAvgPool1d(1)
        
        self.decorder = nn.Sequential(
            nn.Linear(dim_num, size_3),
            nn.GELU(),
            nn.Linear(size_3, size_2),
            nn.GELU(),
            nn.Linear(size_2, size_1)
        ) 

    def forward(self, x):
        x = self.encorder(x)
        x = self.avgpool1d(x)
        x = torch.flatten(x, 1)
        #print(x.size())
        x1 = self.lin(x)
        x = self.decorder(x1)

        return x1, x
```

```python
model = Autoencorder_1d(200, 40)
```

```python
data, true = make_data(n=1000)
data.size(), true.size()
```

```python
out1, out2 = model(data)
out1.size(), out2.size()
```

```python
data.size()
```

```python
model = Autoencorder_1d(200, 5)
optimizer = optim.Adam(params=model.parameters(), lr=1e-3)
criterion = nn.MSELoss()
```

```python
train_data = data[:800, :, :]
val_data = data[800:, : , :]
train_true = true[:800,:]
val_true = true[800:,:]
```

```python
loss_list=[]
val_loss_list = []
for epoch in tqdm(range(5000)):
    model.train()
    index = random.sample([i for i in range(800)], 130)
    
    tmp_train_data = train_data[index, :,:]
    tmp_train_true = train_true[index, :]
    out1, out2 = model(tmp_train_data)
    loss = criterion(out2, tmp_train_true)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    loss_list.append(loss.item())
    
    model.eval()
    out1, out2 = model(val_data)
    loss = criterion(out2, val_true)
    val_loss_list.append(loss.item())
```

```python
plt.figure(figsize=(30,5))
plt.plot(loss_list)
plt.plot(val_loss_list)
plt.show()
```

```python
plt.figure(figsize=(30,5))
plt.plot(loss_list[-50:])
plt.plot(val_loss_list[-50:])
plt.show()
```
